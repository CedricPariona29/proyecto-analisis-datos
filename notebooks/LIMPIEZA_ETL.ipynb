```python

# LIMPIEZA Y TRANSFORMACIÓN ETL
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

# carga del dataset original procesado
df = pd.read_csv('ventas_final.csv')

print("Dataset cargado correctamente")
display(df.head())

print("Dimensiones iniciales:")
print(df.shape)
```

```python

# DIAGNÓSTICO ANTES DEL ETL
filas_iniciales = len(df)
columnas_iniciales = df.shape[1]
duplicados_iniciales = df.duplicated().sum()
nulos_iniciales = df.isnull().sum()

print("Filas iniciales:", filas_iniciales)
print("Columnas iniciales:", columnas_iniciales)
print("Duplicados iniciales:", duplicados_iniciales)

print("\nValores nulos iniciales:")
display(nulos_iniciales)
```

```python

# LIMPIEZA Y TRANSFORMACIÓN ETL
df_etl = df.copy()

# Normalización de nombres de columnas
df_etl.columns = (
    df_etl.columns
    .str.strip()
    .str.lower()
    .str.replace('á', 'a')
    .str.replace('é', 'e')
    .str.replace('í', 'i')
    .str.replace('ó', 'o')
    .str.replace('ú', 'u')
    .str.replace('ñ', 'n')
    .str.replace(' ', '_')
)

# Eliminación de duplicados
df_etl = df_etl.drop_duplicates()

# Tratamiento de valores nulos
df_etl['siaf_o_compra'] = df_etl['siaf_o_compra'].fillna('SIN_REGISTRO')

# Conversión de fechas
df_etl['fecha_o_compra'] = pd.to_datetime(
    df_etl['fecha_o_compra'],
    dayfirst=True,
    errors='coerce'
)

df_etl['fecha_compra'] = pd.to_datetime(
    df_etl['fecha_compra'],
    dayfirst=True,
    errors='coerce'
)

# Conversión de variables numéricas
df_etl['monto'] = pd.to_numeric(df_etl['monto'], errors='coerce')

# Normalización de variables categóricas
columnas_texto = [
    't_orden',
    'nombre_proveedor',
    'descripcion',
    'departamento',
    'provincia',
    'distrito'
]

for col in columnas_texto:
    df_etl[col] = (
        df_etl[col]
        .astype(str)
        .str.strip()
        .str.upper()
    )

# Creación de variable temporal complementaria
df_etl['mes_nombre'] = df_etl['fecha_compra'].dt.month_name()

# Validación posterior al ETL
print("ETL aplicado correctamente.")
print("Dimensiones finales:", df_etl.shape)
print("Duplicados finales:", df_etl.duplicated().sum())

print("\nValores nulos finales:")
display(df_etl.isnull().sum())

print("\nPrimeras filas del dataset transformado:")
display(df_etl.head())
```

**Conclusiones del proceso ETL**

Durante el proceso ETL se realizaron técnicas de limpieza, transformación y normalización de datos con el objetivo de garantizar la calidad y consistencia del dataset utilizado en el análisis.

Transformaciones aplicadas:

- Normalización de nombres de columnas.
- Eliminación de registros duplicados.
- Tratamiento de valores nulos en la variable `siaf_o_compra`.
- Conversión de variables de fecha al formato datetime.
- Conversión de la variable `monto` a formato numérico.
- Estandarización de variables categóricas en mayúsculas.
- Generación de variable temporal complementaria (`mes_nombre`).

Resultados obtenidos:

- El dataset quedó estructurado correctamente para análisis exploratorio y modelado dimensional.
- Se mejoró la consistencia de las variables categóricas.
- Se prepararon variables temporales para análisis de tendencias y KPIs.
- El dataset quedó listo para integración en SQL Server y visualización en Power BI.
