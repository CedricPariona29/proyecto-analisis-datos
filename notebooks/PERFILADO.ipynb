```python
# ============================================================
# PERFILADO Y DIAGNÓSTICO DE CALIDAD
# DESAFÍO COGNITIVO - UAIN1008R
# ============================================================

# Importación de librerías
import pandas as pd
import numpy as np

# Configuración visual
pd.set_option('display.max_columns', None)

# ============================================================
# CARGA DEL DATASET
# ============================================================

df = pd.read_csv('ventas_final.csv')

# Mostrar primeras filas
print("Primeras filas del dataset:")
display(df.head())

# ============================================================
# INFORMACIÓN GENERAL
# ============================================================

print("\nInformación general del dataset:")
df.info()

# ============================================================
# DIMENSIONES DEL DATASET
# ============================================================

print("\nCantidad de filas y columnas:")
print(df.shape)

# ============================================================
# TIPOS DE DATOS
# ============================================================

print("\nTipos de datos:")
display(df.dtypes)

# ============================================================
# VALORES NULOS
# ============================================================

print("\nValores nulos por columna:")
nulos = df.isnull().sum()

display(nulos)

# ============================================================
# PORCENTAJE DE NULOS
# ============================================================

print("\nPorcentaje de valores nulos:")
porcentaje_nulos = (df.isnull().sum() / len(df)) * 100

display(porcentaje_nulos)

# ============================================================
# DUPLICADOS
# ============================================================

print("\nCantidad de filas duplicadas:")
print(df.duplicated().sum())

# ============================================================
# ESTADÍSTICAS DESCRIPTIVAS
# ============================================================

print("\nEstadísticas descriptivas:")
display(df.describe(include='all'))

# ============================================================
# CONCLUSIÓN PRELIMINAR
# ============================================================

print("""
Conclusión preliminar:
El dataset presenta una estructura adecuada para el análisis.
Se identificaron posibles valores nulos y columnas categóricas
que requerirán limpieza y transformación en el proceso ETL.
""")
```

```python
# ============================================================
# VISUALIZACIÓN DE DISTRIBUCIÓN DE MONTO
# ============================================================

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.hist(df['monto'], bins=30)

plt.title('Distribución de Montos')
plt.xlabel('Monto')
plt.ylabel('Frecuencia')

plt.show()

# ============================================================
# TOP 10 PROVEEDORES
# ============================================================

top_proveedores = df['nombre_proveedor'].value_counts().head(10)

plt.figure(figsize=(12,5))

top_proveedores.plot(kind='bar')

plt.title('Top 10 Proveedores con Mayor Número de Registros')
plt.xlabel('Proveedor')
plt.ylabel('Cantidad de Registros')

plt.xticks(rotation=45)

plt.show()

# ============================================================
# DISTRIBUCIÓN TEMPORAL
# ============================================================

ventas_mes = df.groupby('mes_compra')['monto'].sum()

plt.figure(figsize=(10,5))

ventas_mes.plot()

plt.title('Ventas Totales por Mes')
plt.xlabel('Mes')
plt.ylabel('Monto Total')

plt.show()
```
